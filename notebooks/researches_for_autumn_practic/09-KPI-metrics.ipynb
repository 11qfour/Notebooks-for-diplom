```python
import pandas as pd

# Загружаем ОБЩИЙ датасет
try:
    df = pd.read_csv('../data/TRAINING_DATASET.csv')
    print(f"Всего строк: {len(df)}")
except FileNotFoundError:
    df_spec = pd.read_csv('../data/TRAINING_DATASET_SPECIFIC.csv')
    df_gen = pd.read_csv('../data/TRAINING_DATASET_GENERAL.csv')
    df = pd.concat([df_spec, df_gen])

# Очистка от технического шума
noise_categories = ['Неопределено (пусто)']
clean_df = df[~df['label'].isin(noise_categories)].copy()

total_valid_incidents = len(clean_df)
print(f"Всего валидных инцидентов для обработки: {total_valid_incidents}")

# Категории, которые обрабатывает Сортировщик (General)
general_categories = ['Прикладное', 'Инфраструктурное']

# Поток Сортировщика
df_general = clean_df[clean_df['label'].isin(general_categories)]
count_general = len(df_general)

# Поток Специалиста (всё остальное)
df_specific = clean_df[~clean_df['label'].isin(general_categories)]
count_specific = len(df_specific)

# Взвешенный расчет эффективности
ACCURACY_SPECIFIC = 0.9992
ACCURACY_GENERAL = 0.9727  

# Сколько сообщений успешно обработает Специалист
success_specific = count_specific * ACCURACY_SPECIFIC

# Сколько сообщений успешно обработает Сортировщик
success_general = count_general * ACCURACY_GENERAL

# Общий успех
total_success = success_specific + success_general

# Итоговый процент автоматизации
final_kpi = total_success / total_valid_incidents

print("-" * 40)
print(f"Поток 'Специалист' (детальные): {count_specific} шт. (Точность модели: {ACCURACY_SPECIFIC:.2%})")
print(f"Поток 'Сортировщик' (общие):    {count_general} шт. (Точность модели: {ACCURACY_GENERAL:.2%})")
print("-" * 40)
print(f"Успешно автоматизировано (прогноз): {int(total_success)} из {total_valid_incidents}")
print(f"ИТОГОВЫЙ KPI АВТОМАТИЗАЦИИ: {final_kpi:.2%}")
print("-" * 40)

if final_kpi > 0.70:
    print("КРИТЕРИЙ (>70%).")
else:
    print("Критерий не выполнен.")
```
